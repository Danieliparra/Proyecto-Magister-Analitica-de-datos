### Modelo predictivo preliminar

```python
df = pd.read_csv('./df_inicial.csv')

from sklearn.preprocessing import LabelEncoder 
  
for i in df.columns:
  le = LabelEncoder() 
  df_aux = df.copy()
  df_aux[i]= le.fit_transform(df[i])
  df = df_aux
```

```python
random_seed = 44

X= df.drop('readmitted', axis=1)
y= df['readmitted']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=random_seed, shuffle = "True")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 42)

param_grid = {
    'n_estimators': [500],
    'max_depth': [30],
    'criterion': ['gini']
 
}

cv_model = GridSearchCV(estimator = rf, cv=5, scoring='f1_weighted', param_grid = param_grid, verbose=3)

cv_model.fit(X_train, y_train)
print(cv_model.best_params_)
```

```python
mejor_rf = cv_model.best_estimator_
y_pred_rf = mejor_rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

print(classification_report(y_test,y_pred_rf))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf)
```

### Varios Modelos

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
import xgboost as xgb
```

```python
df = pd.read_csv('./df_inicial.csv')

from sklearn.preprocessing import LabelEncoder 
  
for i in df.columns:
  le = LabelEncoder() 
  df_aux = df.copy()
  df_aux[i]= le.fit_transform(df[i])
  df = df_aux
```

```python
from imblearn.over_sampling import RandomOverSampler

random_seed = 44

X= df.drop('readmitted', axis=1)
y= df['readmitted']

oversample = RandomOverSampler(sampling_strategy='minority')

X, y = oversample.fit_resample(X, y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=random_seed, shuffle = "True")
```

```python
clf_gnb = GaussianNB()
clf_gnb = clf_gnb.fit(X_train, y_train)
y_hat_gnb = clf_gnb.predict(X_test)

clf_knn = KNeighborsClassifier(n_neighbors=100)
clf_knn = clf_knn.fit(X_train, y_train)
y_hat_knn = clf_knn.predict(X_test)

clf_dt = DecisionTreeClassifier()
clf_dt = clf_dt.fit(X_train, y_train)
y_hat_dt = clf_dt.predict(X_test)

cld_rf = RandomForestClassifier()
cld_rf = cld_rf.fit(X_train, y_train)
y_hat_rf = cld_rf.predict(X_test)

abc = AdaBoostClassifier()
abc = abc.fit(X_train, y_train)
y_hat_abc = abc.predict(X_test)

xg = xgb.XGBClassifier()
xg = xg.fit(X_train, y_train)
y_hat_xg = xg.predict(X_test)
```

```python
acc_list = []
bacc_list = []
pre_list = []
reca_list = []
f1_list = []
auc_list = []
for pred in [y_hat_gnb, y_hat_knn, y_hat_dt, y_hat_rf, y_hat_abc, y_hat_xg]:
    acc_list.append(accuracy_score(y_test, pred))
    bacc_list.append(balanced_accuracy_score(y_test, pred))
    pre_list.append(precision_score(y_test, pred, average='macro'))
    reca_list.append(recall_score(y_test, pred, average='macro'))
    f1_list.append(f1_score(y_test, pred, average='macro'))

metrics = pd.DataFrame()
models = ['Gaussian Naive Bayes', "k_NN", "Decision Tree", "Random Forest", "AdaBoost", "XGBoost"]

metrics['Model'] = models
metrics['ACC'] = acc_list
metrics['BACC'] = bacc_list
metrics['Precision'] = pre_list
metrics['Recall'] = reca_list
metrics['F1'] = f1_list

metrics.sort_values('F1', ascending=False)
```
